```python
# Happy Mood
df_happy = df[(df['valence'] > 0.55) & (df['energy'] > 0.45) & (df['danceability'] > 0.45) &
              (df['tempo'] >= 85) & (df['mode'] == 1) & (df['loudness'] > -8.5)]

# Sad Mood
df_sad = df[(df['valence'] < 0.45) & (df['energy'] < 0.45) & (df['mode'] == 0) &
            (df['tempo'] < 105) & (df['loudness'] < -7.5) & (df['acousticness'] > 0.35)]

# Angry Mood
df_angry = df[(df['valence'] < 0.55) & (df['energy'] > 0.55) & (df['loudness'] > -7) &
              (df['tempo'] > 95) & (df['speechiness'] > 0.04) & (df['mode'] == 0)]

# Anxious Mood
df_anxious = df[(df['valence'] < 0.45) & (df['energy'] > 0.35) & (df['tempo'] > 105) &
                (df['liveness'] > 0.2) & (df['speechiness'] > 0.04) & (df['mode'] == 0)]

# Relaxed Mood
df_relaxed = df[(df['valence'] > 0.4) & (df['energy'] < 0.5) & (df['acousticness'] > 0.3) &
                (df['instrumentalness'] > 0.05) & (df['tempo'] < 110) & (df['mode'] == 1)]


# Excited Mood
df_excited = df[(df['valence'] > 0.55) & (df['energy'] > 0.55) & (df['danceability'] > 0.55) &
                (df['tempo'] > 105) & (df['loudness'] > -7) & (df['mode'] == 1)]

# Bored Mood
df_bored = df[(df['valence'] < 0.45) & (df['energy'] < 0.35) & (df['tempo'] < 100) &
              (df['danceability'] < 0.45) & (df['speechiness'] < 0.05)]

# Lonely Mood
df_lonely = df[(df['valence'] < 0.45) & (df['energy'] < 0.35) & (df['acousticness'] > 0.4) &
               (df['instrumentalness'] > 0.1) & (df['mode'] == 0) & (df['loudness'] < -7)]

# Hopeful Mood
df_hopeful = df[(df['valence'] > 0.45) & (df['energy'] > 0.35) & (df['mode'] == 1) &
                (df['tempo'] >= 85) & (df['acousticness'] > 0.25)]

# Confident Mood
df_confident = df[(df['valence'] > 0.45) & (df['energy'] > 0.45) & (df['mode'] == 1) &
                  (df['loudness'] > -8) & (df['tempo'] >= 90)]

# Tired Mood
df_tired = df[(df['valence'] < 0.45) & (df['energy'] < 0.35) & (df['tempo'] < 90) &
              (df['acousticness'] > 0.3) & (df['instrumentalness'] > 0.15)]

# Irritated Mood
df_irritated = df[(df['valence'] < 0.55) & (df['energy'] > 0.45) & (df['loudness'] > -7) &
                  (df['speechiness'] > 0.04) & (df['tempo'] > 95)]

# Surprised Mood
df_surprised = df[(df['valence'] > 0.45) & (df['liveness'] > 0.2) & (df['tempo'] > 95) &
                  (df['instrumentalness'] > 0.05) & (df['mode'] == 1)]

# Grateful Mood
df_grateful = df[(df['valence'] > 0.55) & (df['energy'] < 0.45) & (df['acousticness'] > 0.25) &
                 (df['tempo'] >= 85) & (df['mode'] == 1)]

# Fearful Mood
df_fearful = df[(df['valence'] < 0.45) & (df['energy'] > 0.35) & (df['liveness'] > 0.2) &
                (df['mode'] == 0) & (df['tempo'] < 105)]
```

```python
def remove_duplicates(df):
    """
    Removes duplicate rows from the DataFrame based on 'track_name' and 'artist_name'.
    
    Parameters:
    df (pd.DataFrame): The DataFrame from which duplicates should be removed.
    
    Returns:
    pd.DataFrame: A DataFrame with duplicates removed.
    """
    # Remove duplicate rows based on 'track_name' and 'artist_name'
    df_cleaned = df.drop_duplicates(subset=['track_name', 'artist_name'])
    return df_cleaned
```
